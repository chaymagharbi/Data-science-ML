```python
data=pd.read_csv("pima-indians-diabetes.data.csv");
data.columns=["NumTimesPrg","PlGlcConc","BloodP","SkinThick","TwoHourSerIns","BMI","DiPedFunc","age","HasDiabetes"]
data.head(10)
```

```python
data.info()
```

```python
import numpy as np;
cols_with_missing = ["BloodP", "SkinThick", "TwoHourSerIns", "BMI"]

for col in cols_with_missing:
    non_zero = data[col][data[col] != 0]
    med = non_zero.median()
    data.loc[data[col] == 0, col] = med
```

```python
print(data);
```

```python
df_min = data.min()
df_max = data.max()

df_norm = (data - df_min) / (df_max - df_min)
print(df_norm)
```

```python
num_cols = data.select_dtypes(include=['float64','int64']).columns
```

```python
df_mean = data[num_cols].mean()
df_std = data[num_cols].std()

data_standard = data.copy()
data_standard[num_cols] = (data[num_cols] - df_mean) / df_std
print(data_standard)
```
